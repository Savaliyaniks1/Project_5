```python
def remove_outliers(col):
  sorted(col)
  q1,q3 = np.percentile(col,[25,75])
  iqr = q3-q1
  l1 = q1-(1.5*iqr)
  u1 = q3+(1.5*iqr)
  return l1,u1
```

```python
for i in xdr_data[
  [
      'MSISDN/Number',
      'Total UL (Bytes)',
      'Dur. (ms)'

  ]
]:
  if xdr_data[i].dtype != 'object':
    l1,u1 = remove_outliers(xdr_data[i])
    xdr_data[i] = np.where(xdr_data[i] > u1, u1, xdr_data[i])
    xdr_data[i] = np.where(xdr_data[i] < l1, l1, xdr_data[i])
```

# Variable transformations

```python
# Calculate total duration for all sessions per user
user_total_duration = xdr_data.groupby('MSISDN/Number')['Dur. (ms)'].sum().reset_index()

# Calculate deciles for total duration
# Set duplicates='drop' to handle non-unique bin edges
user_total_duration['Decile'] = pd.qcut(user_total_duration['Dur. (ms)'], 10, labels=False, duplicates='drop') + 1

# Select top 5 deciles
top_5_deciles = user_total_duration[user_total_duration['Decile'] <= 5]

# Merge with original data to get total data per user
merged_data = pd.merge(xdr_data, top_5_deciles[['MSISDN/Number', 'Decile']], on='MSISDN/Number', how='inner')

# Calculate total data (DL+UL) per decile class
total_data_per_decile = merged_data.groupby('Decile').agg({
    'Total DL (Bytes)': 'sum',
    'Total UL (Bytes)': 'sum'
})

# Calculate total data (DL+UL)
total_data_per_decile['Total Data (Bytes)'] = total_data_per_decile['Total DL (Bytes)'] + total_data_per_decile['Total UL (Bytes)']

print(total_data_per_decile)
```

```python
# prompt: 	Dimensionality Reduction  perform a principal component analysis to reduce the dimensions of your data and provide a useful interpretation of the results (Provide your interpretation in four (4) bullet points maximum).

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Select numerical features for PCA
numeric_features = ['Dur. (ms)', 'Total DL (Bytes)', 'Total UL (Bytes)']
X = xdr_data[numeric_features]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 components for visualization
principal_components = pca.fit_transform(X_scaled)

# Create a DataFrame with principal components
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Explained Variance Ratio: {explained_variance_ratio}")
```
